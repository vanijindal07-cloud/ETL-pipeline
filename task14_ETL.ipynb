```python
import pandas as pd

df = pd.read_csv(
    'Amazon Sale Report.csv',
    engine='python',
    encoding='latin1',
    sep=',',
    quoting=3,   # ignore quotes issues
    on_bad_lines='skip'
)

df.head()
```

```python
df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_')
```

```python
import os

os.makedirs('raw', exist_ok=True)
os.makedirs('processed', exist_ok=True)
os.makedirs('output', exist_ok=True)
```

```python
os.listdir()
```

```python
df.to_csv('raw/raw_data.csv', index=False)
```

```python
df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_')
```

```python
df = df.drop_duplicates()
```

```python
df.isnull().sum()
```

```python
df['amount'] = df['amount'].fillna(0)
df['currency'] = df['currency'].fillna('INR')
```

```python
df['date'] = pd.to_datetime(df['date'], errors='coerce')
df['amount'] = pd.to_numeric(df['amount'], errors='coerce')
```

```python
df['profit'] = df['amount'] * 0.2
```

```python
df['high_value_order'] = df['amount'] > 500
```

```python
df['month'] = df['date'].dt.month
```

```python
df.to_csv('processed/cleaned_data.csv', index=False)
```

```python
orders = df[['order_id','date','status','amount','ship_city','ship_state']]
```

```python
print(df.columns)
```

```python
orders = df[['order_id','date','status','amount','ship-city','ship-state']]
```

```python
df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_').str.replace('-', '_')
```

```python
orders = df[['order_id','date','status','amount','ship_city','ship_state']]
```

```python
products = df[['sku','category','size','asin']].drop_duplicates()
```

```python
customers = df[['ship_city','ship_state','ship_postal_code']].drop_duplicates()
```

```python
orders.to_csv('output/orders.csv', index=False)
products.to_csv('output/products.csv', index=False)
customers.to_csv('output/customers.csv', index=False)
```

```python
import os
os.listdir('output')
```

```python
import sqlite3

conn = sqlite3.connect('output/amazon_sales.db')
```

```python
orders.to_sql('orders', conn, if_exists='replace', index=False)
products.to_sql('products', conn, if_exists='replace', index=False)
customers.to_sql('customers', conn, if_exists='replace', index=False)
```

```python
conn.close()
```
